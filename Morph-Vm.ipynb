```python
# Estimate k using Silhouette Coefficient - Intercluster similarity
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
Scores = []  # 
for k in range(2,9):
    estimator = AgglomerativeClustering(n_clusters=k,linkage='average')  # 
    estimator.fit(fmr)
    Scores.append(silhouette_score(fmr,estimator.labels_,metric='euclidean'))
X = range(2,9)
plt.figure()
plt.xlabel('k')
plt.ylabel('Silhouette Coefficient')
plt.plot(X,Scores,'o-')
plt.title('Silhoutte')
#plt.savefig('k_estimitor.png')
plt.show()
```

```python
from sklearn.decomposition import PCA
import palettable
#fmr=fmr.iloc[:,1:]
clustering = AgglomerativeClustering(n_clusters=5).fit(fmr)

pca = PCA(n_components=0.6)
new_pca = pd.DataFrame(pca.fit_transform(fmr))
importance = pca.explained_variance_ratio_


c=pca.components_

#print('Coefficients Matrix')
print(importance)
pca.components_=pd.DataFrame(pca.components_)
#pca.components_.columns=fb.columns
pca.components_.index=['1st','2nd','3rd','4th','5th']

plt.figure(figsize=(20,8))
sns.heatmap(round(pca.components_,3),annot=True,cmap=palettable.cartocolors.diverging.ArmyRose_7.mpl_colors)
plt.xlabel('Feature',fontsize=20)
plt.ylabel('Principal Component',fontsize=20)
```

```python
clustering = AgglomerativeClustering(n_clusters=3).fit(fmr)

pca = PCA(n_components=0.6)
new_pca = pd.DataFrame(pca.fit_transform(fmr))
importance = pca.explained_variance_ratio_


c=pca.components_
output_data = pd.concat([ pd.Series(clustering.labels_, index=fmr.index),fmr], axis=1)   # Output with group information
output_data.columns = ['Type']+list(fmr.columns)  # Rename columns


t0=new_pca[output_data['Type']==0]
t1=new_pca[output_data['Type']==1]
t2=new_pca[output_data['Type']==2]

fig=plt.figure()
ax=fig.add_subplot(111)
ax.scatter(t0.iloc[:,0],t0.iloc[:,1],marker='.',label='Type1')
ax.scatter(t1.iloc[:,0],t1.iloc[:,1],marker='.',label='Type2')
ax.scatter(t2.iloc[:,0],t2.iloc[:,1],marker='.',label='Type3')

kwargs={'linestyle':'-', 'lw':0.1, 'width':0.02} 
for i in range(c.shape[1]):
    ax.arrow(0,0,c[0,i],c[1,i]-0.02,alpha=0.5,ec='green',fc='green',**kwargs)
    ax.text(c[0,i],c[1,i],f'{i+1}', ha="center", va="center")
ax.set_xlabel(f'PC1:{round(importance[0]*100,1)}%',fontsize=14)  
ax.set_ylabel(f'PC2:{round(importance[1]*100,1)}%',fontsize=14)
```

```python
from sklearn.model_selection import StratifiedKFold, KFold
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn import svm

X=mb
y=fb.iloc[:,0]

clf_svm=svm.SVC(kernel='rbf')
scores_svm = cross_val_score(clf_svm, X, y, cv=10)
print("SVM: %0.2f accuracy with a standard deviation of %0.2f" % (scores_svm.mean(), scores_svm.std()))
s_svm=scores_svm
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import plot_confusion_matrix

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Run classifier, using a model that is too regularized (C too low) to see
# the impact on the results
classifier = svm.SVC(kernel='linear').fit(X_train, y_train)
class_names = ['Type1','Type2','Type3','Type4','Type5']
# Plot non-normalized confusion matrix
titles_options = [("Confusion matrix, without normalization", None),
                  ("Normalized confusion matrix", 'true')]
for title, normalize in titles_options:
    disp = plot_confusion_matrix(classifier, X_test, y_test,
                                 display_labels=class_names,
                                 cmap=plt.cm.Blues,
                                 normalize=normalize)
    disp.ax_.set_title(title)

    print(title)
    print(disp.confusion_matrix)

plt.show()
```
